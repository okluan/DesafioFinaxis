```python
ano_inicial = int (input("Digite o ano inicial: "))

periodo_inicial = int (input("Digite o mês de inicio: "))

ano_final = int(input("Digite o ano final: "))

periodo_final  = int (input("Digite o mês final: "))

#computando anos

#lista de anos
anos = []
anos.append(ano_inicial)
while ano_inicial != ano_final:
    ano = ano_inicial + 1
    ano_inicial = ano_inicial + 1
    anos.append(ano)

# ano inicial
meses_ano_inicial = []
meses_ano_inicial.append (periodo_inicial)
i = 1
while i != 12:
    i = i + 1
    if i not in meses_ano_inicial:
        meses_ano_inicial.append(i)


ano_inicio = anos[0]
anos.pop(0)

# ano final
meses_ano_final = []
meses_ano_inicial.append(periodo_final)
i = 0
while i != periodo_final:
    i = i + 1
    if i not in meses_ano_final:
        meses_ano_final.append(i)

ano_final = anos[-1]
anos.pop(-1)


#concatenando meses e ano inicial
meses_ano_inicial_str =[]
for i in meses_ano_inicial:
    if i <10:
        num = '0'+ str(i)
        meses_ano_inicial
        meses_ano_inicial_str.append(num)
    else: 
        num =  str(i)
        meses_ano_inicial_str.append(num)

meses_e_ano_inicial = []       
for i in meses_ano_inicial_str:
    num = str(ano_inicio) + i
    meses_e_ano_inicial.append(num)

    
#concatenando meses e ano final
meses_ano_final_str =[]
for i in meses_ano_final:
    if i <10:
        num = '0'+ str(i)
        meses_ano_final_str.append(num)
    else: 
        num =  str(i)
        meses_ano_final_str.append(num)

meses_e_ano_final = []       
for i in meses_ano_final_str:
    num = str(ano_final) + i
    meses_e_ano_final.append(num)
```
